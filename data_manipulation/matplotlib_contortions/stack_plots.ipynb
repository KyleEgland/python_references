```python
# NOTE: The first line, "%matplotlib inline" is necessary only for the Jupyter notebook. You won't
# need it in a "regular" .py file.
%matplotlib inline
import matplotlib.pyplot as plt

# We're going to make a stack plot of what a persons day consists of, there are the days of the week (numbers),
# then the activities - each given in hours
days = [1, 2, 3, 4, 5]
sleeping = [7, 8, 6, 11, 7]
eating = [2, 3, 4, 3, 2]
working = [7, 8, 7, 2, 2]
playing = [8, 5, 7, 8, 13]

# Creating the stack plot:
# First is the x-axis (in this case the day of the week)
# After that is all of the y elements which will be stacked in order (bottom to top)
plt.stackplot(days, sleeping, eating, working, playing, colors=['m', 'c', 'r', 'k'],
              labels=['Sleeping', 'Eating', 'Working', 'Playing'])

# NOTE: labels can also be achieved by using an empty plot, this isn't necessary but was a part
# of the video tutorial that I found intriguing, may be useful for later?
# plt.plot([], [], color='m', label='Sleeping', linewidth=5)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Days Stack Plot')
plt.legend()
plt.show()
```
