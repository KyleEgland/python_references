Generally, scatter plots are used for things like visuallizing a distribution of data

```python
# NOTE: The first line, "%matplotlib inline" is necessary only for the Jupyter notebook. You won't
# need it in a "regular" .py file.
%matplotlib inline
import matplotlib.pyplot as plt


# This is probably bad data to demonstrate with because it's arbitrary and will not
# have a clear correlation <- remember, I'm doing this for demonstration purposes...
x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [5, 2, 4, 2, 1, 4, 5, 2]

# Marker on scatter plot is type of "dot" used to represent each point, s represents the size of the "dot"
# check https://matplotlib.org/ for more information
plt.scatter(x, y, label='Skattered', color='k', marker='*', s=120)

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('This is a Scatter Plot\nNeat Isn\'t it?')
```
