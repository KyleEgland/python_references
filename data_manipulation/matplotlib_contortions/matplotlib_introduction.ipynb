```python
# Normally (as in "unwritten convention") matplotlib.pyplot (the thing that does the plotting) is
# imported as plt.
# NOTE: The first line, "%matplotlib inline" is necessary only for the Jupyter notebook. You won't
# need it in a "regular" .py file.
%matplotlib inline
import matplotlib.pyplot as plt


# Plotting some stuff; plt.plot(x, y) <- the order in which things are plotted
plt.plot([1, 2, 3], [5, 7, 4])

# The above line does plotting but it's in the background - nothing will display.
# So, we use the .show() method to display what we've done
plt.show()
```

```python
# Labeling stuff

# Setting up the things we'll plot
x = [1, 2, 3]
y = [5, 7, 4]
# Creating second plot vars to demonstrate the legend
x2 = [1, 2, 3]
y2 = [10, 14, 12]

# Creating the plot
plt.plot(x, y, label='First Line')
plt.plot(x2, y2, label='Second Line')

# Labels can be inserted anytime before the .show() method is invoked
plt.xlabel('Plot Number')
plt.ylabel('Important var')

# Add title to plot
plt.title('Interesting Graph\nCheck it Out')

# Adding a legend. This will use the "label" attribute from the .plot above
plt.legend()

plt.show()
```
