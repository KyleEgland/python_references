### Loading a CSV File
This is the first part of the tutorial/demo which uses the Python built in module "CSV"

```python
# NOTE: The first line, "%matplotlib inline" is necessary only for the Jupyter notebook. You won't
# need it in a "regular" .py file.
%matplotlib inline
import matplotlib.pyplot as plt
import csv

# Instantiate some lists to contain our values from file
x = []
y = []

# Read in the target file
with open('example_data.txt', 'r') as csvFile:
    # Store the stream in a variable using csv read to parse it,
    # delimiter tells csv module what to look for seperating the
    # values in each row (not the rows themselves)
    plots = csv.reader(csvFile, delimiter=',')
    # Loop through each row...
    for row in plots:
        # Put the first value in the row into our x list
        x.append(int(row[0]))
        # Put the second value in the row into our y list
        y.append(int(row[1]))

plt.plot(x, y, label='File Contents')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Loaded from File\nCSV')
plt.legend()
plt.show()
```

### Loading File with Numpy
This is the second part of the tutorial which will be using the "Numpy" module to import our file

```python
import numpy as np


# Using numpy we'll "unpack" our values using the .loadtxt() method <- does NOT have to be a txt file
x, y = np.loadtxt('example_data.txt', delimiter=',', unpack=True)

plt.plot(x, y, label='File Contents')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Loaded from File\nNumpy')
plt.legend()
plt.show()
```
