```python
# NOTE: The first line, "%matplotlib inline" is necessary only for the Jupyter notebook. You won't
# need it in a "regular" .py file.
%matplotlib inline
import matplotlib.pyplot as plt

slices = [7, 2, 2, 13]
activities = ['Sleeping', 'Eating', 'Working', 'Playing']
colors = ['c', 'm', 'r', 'b']

# starting angle specifies where the "top" is, notice that "playing" starts at the 12 o'clock position
# explode will "pull out" a piece of the pie, in this case 0.1 will grab the second item in our list (eating).
# Each digit in the tuple represents an element in the pie chart, so changing the first 0 to something else will
# pull the "sleeping" slice out
# autopct='%1.1f%%' will add percentages to the pie <- Uses string formatting that I don't quite understand, though
# I have figured out that the second digit will change the decimal place
plt.pie(slices, labels=activities, colors=colors, startangle=90, shadow=True,
       explode=(0, 0.1, 0, 0), autopct='%1.1f%%')

plt.title('A Day in the Life')
# The legend covers up part of the chart so I left it out, uncomment it to see what I'm talking about
# plt.legend()
plt.show()
```
